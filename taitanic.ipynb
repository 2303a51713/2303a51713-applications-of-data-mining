```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
df['Age'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
```
